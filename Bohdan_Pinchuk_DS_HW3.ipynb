## –ó–∞–≤–¥–∞–Ω–Ω—è 6. –ë–∞–∑–æ–≤–∞ —Ñ—É–Ω–∫—Ü—ñ—è + —Ü–∏–∫–ª
**–£–º–æ–≤–∞**

–ù–∞–ø–∏—à—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é ```count_positive(numbers)```, —è–∫–∞ –ø—Ä–∏–π–º–∞—î —Å–ø–∏—Å–æ–∫ —á–∏—Å–µ–ª —ñ –ø–æ–≤–µ—Ä—Ç–∞—î –∫—ñ–ª—å–∫—ñ—Å—Ç—å –¥–æ–¥–∞—Ç–Ω–∏—Ö —á–∏—Å–µ–ª —É —Å–ø–∏—Å–∫—É.

**–í–∏–º–æ–≥–∏:**
* –≤–∏–∫–æ—Ä–∏—Å—Ç–∞—Ç–∏ —Ü–∏–∫–ª for
* –≤–∏–∫–æ—Ä–∏—Å—Ç–∞—Ç–∏ return
* –Ω–µ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–≤–∞—Ç–∏ print –≤—Å–µ—Ä–µ–¥–∏–Ω—ñ —Ñ—É–Ω–∫—Ü—ñ—ó

````python
#### Variant ‚Ññ1 ####

import random as rnd

# Input data
lower = -10
upper = 10
count_nums = 10

# Solution
numbers = [rnd.randrange(lower, upper + 1) for i in range(count_nums)]

def count_positive(numbers):
    """
    Calculate the count of the positive numbers in a list of numbers.
    :param numbers: list of numbers
    :return: count of positive numbers
    """
    if numbers == [] or len(numbers) == 0:
        raise ValueError("Input array is empty!")
    else:
        count = 0
        for number in numbers:
            if number > 0:
                count += 1
        return count


# Print results
print(f"–í—Ö—ñ–¥–Ω–∏–π –º–∞—Å–∏–≤ –¥–∞–Ω–∏—Ö: {numbers}")
print(f"–ö—ñ–ª—å–∫—ñ—Å—Ç—å –¥–æ–¥–∞—Ç–Ω–∏—Ö —á–∏—Å–µ–ª —É —Å–ø–∏—Å–∫—É: {count_positive(numbers)}")
````

–ù–∞–ø–∏—à—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é ```number_type(n)```, —è–∫–∞:
* –ø–æ–≤–µ—Ä—Ç–∞—î ```"positive"```, —è–∫—â–æ —á–∏—Å–ª–æ –¥–æ–¥–∞—Ç–Ω–µ
* ```"negative"```, —è–∫—â–æ –≤—ñ–¥ º—î–º–Ω–µ
* ```"zero"```, —è–∫—â–æ –¥–æ—Ä—ñ–≤–Ω—é—î –Ω—É–ª—é

**–í–∏–º–æ–≥–∏:**
* –≤–∏–∫–æ—Ä–∏—Å—Ç–∞—Ç–∏ –∫—ñ–ª—å–∫–∞ ```return```
* –Ω–µ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–≤–∞—Ç–∏ –≤–∫–ª–∞–¥–µ–Ω—ñ ```if```

````python
#### Variant ‚Ññ1 ####

import random as rnd

# Input data
lower = -10
upper = 10
number = rnd.randrange(lower, upper + 1)

# Solution
def number_type(n):
    """
    Return the type of number.
    :param n: number
    :return: type of number
    """
    if n < 0:
        return "negative"
    elif n > 0:
        return "positive"
    else:
        return "zero"

# Print results
print(f"–ß–∏—Å–ª–æ \"{number}\" –º–∞—î —Ç–∏–ø \"{number_type(number)}\"")
````
