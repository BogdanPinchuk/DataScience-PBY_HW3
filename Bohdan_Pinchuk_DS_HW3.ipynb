## –ó–∞–≤–¥–∞–Ω–Ω—è 6. –ë–∞–∑–æ–≤–∞ —Ñ—É–Ω–∫—Ü—ñ—è + —Ü–∏–∫–ª
**–£–º–æ–≤–∞**

–ù–∞–ø–∏—à—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é ```count_positive(numbers)```, —è–∫–∞ –ø—Ä–∏–π–º–∞—î —Å–ø–∏—Å–æ–∫ —á–∏—Å–µ–ª —ñ –ø–æ–≤–µ—Ä—Ç–∞—î –∫—ñ–ª—å–∫—ñ—Å—Ç—å –¥–æ–¥–∞—Ç–Ω–∏—Ö —á–∏—Å–µ–ª —É —Å–ø–∏—Å–∫—É.

**–í–∏–º–æ–≥–∏:**
* –≤–∏–∫–æ—Ä–∏—Å—Ç–∞—Ç–∏ —Ü–∏–∫–ª for
* –≤–∏–∫–æ—Ä–∏—Å—Ç–∞—Ç–∏ return
* –Ω–µ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–≤–∞—Ç–∏ print –≤—Å–µ—Ä–µ–¥–∏–Ω—ñ —Ñ—É–Ω–∫—Ü—ñ—ó

````python
#### Variant ‚Ññ1 ####

import random as rnd

# Input data
lower = -10
upper = 10
count_nums = 10

# Solution
numbers = [rnd.randrange(lower, upper + 1) for _ in range(count_nums)]


def count_positive(numbers):
    """
    Calculate the count of the positive numbers in a list of numbers.
    :param numbers: list of numbers
    :return: count of positive numbers
    """
    if numbers == [] or len(numbers) == 0:
        raise ValueError("Input array is empty!")
    else:
        count = 0
        for number in numbers:
            if number > 0:
                count += 1
        return count


# Print results
print(f"–í—Ö—ñ–¥–Ω–∏–π –º–∞—Å–∏–≤ –¥–∞–Ω–∏—Ö: {numbers}")
print(f"–ö—ñ–ª—å–∫—ñ—Å—Ç—å –¥–æ–¥–∞—Ç–Ω–∏—Ö —á–∏—Å–µ–ª —É —Å–ø–∏—Å–∫—É: {count_positive(numbers)}")
````

## –ó–∞–≤–¥–∞–Ω–Ω—è 8. –§—É–Ω–∫—Ü—ñ—è + —Ü–∏–∫–ª + —É–º–æ–≤–∞
**–£–º–æ–≤–∞**

–ù–∞–ø–∏—à—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é ```max_in_list(numbers)```, —è–∫–∞ –ø–æ–≤–µ—Ä—Ç–∞—î –Ω–∞–π–±—ñ–ª—å—à–µ —á–∏—Å–ª–æ –∑—ñ —Å–ø–∏—Å–∫—É.

**–í–∏–º–æ–≥–∏:**
* –≤–∏–∫–æ—Ä–∏—Å—Ç–∞—Ç–∏ —Ü–∏–∫–ª
* –Ω–µ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–≤–∞—Ç–∏ ```max()```
* —è–∫—â–æ —Å–ø–∏—Å–æ–∫ –ø–æ—Ä–æ–∂–Ω—ñ–π ‚Äî –ø–æ–≤–µ—Ä–Ω—É—Ç–∏ ```None```

````python
#### Variant ‚Ññ1 ####

import random as rnd

# Input data
lower = -100
upper = 100

# Solution
count_nums = rnd.randint(0, 11)
numbers = [rnd.randrange(lower, upper + 1) for _ in range(count_nums)]


def max_in_list(numbers):
    """
    Return the maximum number in a list of numbers.
    :param numbers: list of numbers
    :return: maximum number
    """
    if numbers == [] or len(numbers) == 0:
        return None
    else:
        max_number = numbers[0]
        for index in range(1, len(numbers)):
            if numbers[index] > max_number:
                max_number = numbers[index]
        return max_number


report = max_in_list(numbers)

# Print results
print(f"–í—Ö—ñ–¥–Ω–∏–π –º–∞—Å–∏–≤ –¥–∞–Ω–∏—Ö: {numbers}")
print(f"–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–µ –∑–Ω–∞—á–µ–Ω–Ω—è —ñ–∑ —Å–ø–∏—Å–∫—É: {report}")
````

## –ó–∞–≤–¥–∞–Ω–Ω—è 9. –û–±—Ä–æ–±–∫–∞ –ø–æ–º–∏–ª–æ–∫
**–£–º–æ–≤–∞**

–ù–∞–ø–∏—à—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é ```f(x)```, —è–∫–∞ –æ–±—á–∏—Å–ª—é—î $f(x)=\sqrt{\frac{1}{1+x}}$.

**–í–∏–º–æ–≥–∏:**
* –≤–∏–∫–æ—Ä–∏—Å—Ç–∞—Ç–∏ ```try ... except```
* –û–±—Ä–æ–±–∏—Ç–∏ ```ZeroDivisionError``` —Ç–∞ ```ValueError```.

–ê–Ω–∞–ª—ñ–∑ –¥–æ–ø—É—Å—Ç–∏–º–∏—Ö –∑–Ω–∞—á–µ–Ω—å —á–∏—Å–ª–∞ $x$ (–±–µ–∑ –≤—Ä–∞—Ö—É–≤–∞–Ω–Ω—è –∫–æ–º–ø–ª–µ–∫—Å–Ω–∏—Ö —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ñ–≤).
–†–æ–∑–≤ º—è–∂–µ–º–æ –Ω–µ—Ä—ñ–≤–Ω—ñ—Å—Ç—å:
$$\frac{1}{1+x}\geq 0,$$
$$\begin{matrix}
{1+x}>0, & {1+x}\neq 0,\\
\end{matrix}$$
$$\begin{matrix}
x>-1, & x\neq -1.\\
\end{matrix}$$



````python
#### Variant ‚Ññ1 ####

import random as rnd
from math import sqrt

# Input data
lower = -10
upper = 10
tolerance = 4

# Solution
value = rnd.randrange(lower, upper + 1)


def f(x):
    """
    Calculate the f(x) = sqrt(1/(1+x)) function.
    :param x: input value, requirement: x > -1
    :return: result of f(x)
    """
    try:
        result = sqrt(1.0 / (1 + x))
        return result
    except ZeroDivisionError:
        print(f"–ó–Ω–∞—á–µ–Ω–Ω—è \"{x}\" –ø—Ä–∏–∑–≤–æ–¥–∏—Ç—å –¥–æ –¥—ñ–ª–µ–Ω–Ω—è –Ω–∞ –Ω—É–ª—å!")
    except ValueError:
        print(f"–ó–Ω–∞—á–µ–Ω–Ω—è \"{x}\" –ø—Ä–∏–∑–≤–æ–¥–∏—Ç—å –¥–æ –∫–æ–º–ø–ª–µ–∫—Å–Ω–æ–≥–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—É!")


report = f(value)

# Print results
if report is not None:
    print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—É —Ñ—É–Ω–∫—Ü—ñ—ó: f({value}) = {report:.{tolerance}f}")
````

## –ó–∞–≤–¥–∞–Ω–Ω—è 10. –õ—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—è

–î–∞–Ω–æ —Å–ø–∏—Å–æ–∫ —á–∏—Å–µ–ª:

```
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
```

–°—Ç–≤–æ—Ä—ñ—Ç—å –ª—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—é, —è–∫–∞:
* –ø—Ä–∏–π–º–∞—î –æ–¥–Ω–µ —á–∏—Å–ª–æ
* –ø–æ–≤–µ—Ä—Ç–∞—î ```True```, —è–∫—â–æ —á–∏—Å–ª–æ –ø–∞—Ä–Ω–µ
* ```False``` ‚Äî —è–∫—â–æ –Ω–µ–ø–∞—Ä–Ω–µ

–ü–æ—Ç—ñ–º –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–π—Ç–µ —Ü—é –ª—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—é, —â–æ–± –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é list comprehension —Å—Ç–≤–æ—Ä–∏—Ç–∏ –Ω–æ–≤–∏–π —Å–ø–∏—Å–æ–∫ —Ç—ñ–ª—å–∫–∏ —ñ–∑ –ø–∞—Ä–Ω–∏–º–∏ —á–∏—Å–ª–∞–º–∏.

````python
#### Variant ‚Ññ2 ####

# Input data
numbers = [1, 2, 3, 4, 5, 6, 7, 8]

# Solution
# Stage ‚Ññ1
is_even = lambda x: x % 2 == 0
input_list = [is_even(number) for number in numbers]
dictionary = dict(zip(numbers, input_list))

srt_col1 = 'Number'
srt_col2 = ' Is even'
max_len_lf = len(srt_col1)
max_len_rt = len(srt_col2)

for key, value in dictionary.items():
    max_len_lf = max(max_len_lf, len(f"{key}"))
    max_len_rt = max(max_len_rt, len(f"{value}"))

report = f"{f"{srt_col1}":<{max_len_lf + 1}}|{f"{srt_col2}":^{max_len_rt + 1}}"
report += f"\n{'-':-^{max_len_lf + max_len_rt + 2}}"
for key, value in dictionary.items():
    report += f"\n{key:<{max_len_lf + 1}}| {f"{value}":<{max_len_rt + 1}}"
report += f"\n{'-':-^{max_len_lf + max_len_rt + 2}}"

# Print results
print(report)

# Stage ‚Ññ2
output_list = [key for key, value in dictionary.items() if value]

# Print results
print(f"–°–ø–∏—Å–æ–∫ —Ç—ñ–ª—å–∫–∏ –ø–∞—Ä–Ω–∏—Ö —á–∏—Å–µ–ª: {output_list}")
````
